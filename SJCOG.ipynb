Here’s a **concise one-sentence explanation** for each attribute:

1. **Block Group 2010** – Unique 2010 census block group identifier.  
2. **Block Group 2020** – Updated 2020 census block group identifier.  
3. **State Code** – Numeric code representing the U.S. state.  
4. **County Code** – Numeric code representing the county within the state.  
5. **Tract Code** – Census tract identifier within the county.  
6. **Block Group Code** – Final digit identifying the block group within a tract.  
7. **CSA Code** – Code for the Combined Statistical Area (CSA), if applicable.  
8. **CSA Name** – Name of the Combined Statistical Area (e.g., large metro regions).  
9. **CBSA Code** – Code for the Core-Based Statistical Area (metro/micro area).  
10. **CBSA Name** – Name of the CBSA, typically a metropolitan or micropolitan area.  
11. **Workers** – Number of employed people living in the block group.  
12. **Employment Entropy** – Measure of job type diversity in the area (higher = more diverse).  
13. **Employment Household Entropy** – Combined diversity of job types and household types.  
14. **Intersection Density** – Number of street intersections per square kilometer (walkability proxy).  
15. **Distance to Transit** – Average distance to nearest public transit stop (in meters or feet).  
16. **Employment Household Entropy Rank** – Relative ranking of combined job and household diversity.  
17. **Employment Entropy Rank** – Relative ranking of job diversity.  
18. **Intersection Density Rank** – Ranking based on intersection density (more intersections = more walkable).  
19. **Distance to Transit Rank** – Rank showing how close the area is to transit options (lower = closer).  
20. **Walkability** – Composite score representing how walkable the area is based on urban form and access.

# VMT Estimation Approach

In this project, Vehicle Miles Traveled (VMT) was estimated using person-level trip data from the Replica dataset, with a specific focus on commute-related VMT. While VMT is traditionally defined as the total vehicle miles traveled across all trip purposes (e.g., work, school, errands, leisure), our analysis centers on work trips only, consistent with the project's objective to understand and reduce commuter travel demand in San Joaquin County. To ensure an accurate estimation of vehicle movement and avoid double-counting, we adopted a filtering method that includes only trips where the primary_mode was private_auto, representing the vehicle driver. Trips labeled as auto_passenger were excluded, as these represent passengers in shared vehicles whose travel is already captured through the corresponding driver's trip. This method aligns with established best practices in transportation planning and emissions analysis by ensuring that each vehicle trip is counted only once, producing a reliable and policy-relevant measure of commute-based VMT

```python
# filtering commute trips only (work purpose)
df_commute = df_all[df_all['trip_purpose'] == 'work']
```

```python
vehicle_driver_modes = ['private_auto']
df_vmt = df_commute[df_commute['primary_mode'].isin(vehicle_driver_modes)]
```

```python
vmt_summary = df_vmt.groupby('trip_category').agg(
    total_vmt_miles=('trip_distance_miles', 'sum'),
    avg_trip_miles=('trip_distance_miles', 'mean'),
    avg_trip_duration_min=('trip_duration_minutes', 'mean'),
    trip_count=('trip_distance_miles', 'count')
).reset_index()
vmt_summary
```

```python
df_vmt['is_sov'] = df_vmt['primary_mode'] == 'private_auto'

sov_share = df_vmt.groupby('trip_category')['is_sov'].mean().reset_index(name='sov_share')
sov_share
```

# GHG Emission Estimation

To estimate greenhouse gas (GHG) emissions associated with commuter travel, we focused on trips made by private automobiles using internal combustion engines. Using the filtered Replica dataset containing only private_auto trips, we excluded all trips where the vehicle_fuel_type was labeled as electric to ensure that emissions estimates reflect only tailpipe-emitting vehicles. We then applied the U.S. EPA’s standard emission factor of 404 grams of CO₂ per mile (or 0.404 kg/mile) to the total vehicle miles traveled (VMT) for each trip category. This method provides a realistic approximation of GHG emissions generated specifically by gas- and diesel-powered commute trips, aligning with best practices for emissions modeling and ensuring that our sustainability analysis accurately reflects vehicle-based environmental impacts.

```python
# Filter out EVs — these don't emit tailpipe GHGs
df_vmt_gas = df_vmt[df_vmt['vehicle_fuel_type'] != 'electric']
```

```python
EMISSION_FACTOR_KG_MILE = 0.404

# Group by trip category
ghg_summary = df_vmt_gas.groupby('trip_category').agg(
    total_vmt_miles=('trip_distance_miles', 'sum')
).reset_index()

# Add emissions
ghg_summary['total_ghg_kg'] = ghg_summary['total_vmt_miles'] * EMISSION_FACTOR_KG_MILE
ghg_summary
```

# Intra County Commute Single Occupancy Vehicle Share calculation

```python
#Filter for Intra-County Commute Trips

df_intra = df_commute[df_commute['trip_category'] == 'Intra-county']
```

```python
#Calculate SOV Share

# Total trips (for mode share, not VMT)
total_intra_trips = df_intra.shape[0]

# SOV count
sov_count = df_intra[df_intra['primary_mode'] == 'private_auto'].shape[0]

# SOV share
sov_share = sov_count / total_intra_trips

print(f"SOV Share (Intra-county commute trips): {sov_share:.2%}")
```

```python
df_commute['primary_mode'].value_counts()
```

```python
alt_modes = ['walking', 'public_transit', 'biking']
df_alt = df_intra[df_intra['primary_mode'].isin(alt_modes)]
```

```python
tract_summary = df_intra.groupby('origin_trct_fips_2020').agg(
    total_trips=('activity_id', 'count'),
    sov_trips=('primary_mode', lambda x: (x == 'private_auto').sum()),
    alt_trips=('primary_mode', lambda x: x.isin(alt_modes).sum())
)

tract_summary['sov_share'] = tract_summary['sov_trips'] / tract_summary['total_trips']
tract_summary['alt_share'] = tract_summary['alt_trips'] / tract_summary['total_trips']
tract_summary
```

# Commute Mode Share by Trip Type

```python
import matplotlib.pyplot as plt

mode_share = df_commute.groupby(['trip_category', 'primary_mode']).size().unstack(fill_value=0)
mode_share_percent = mode_share.div(mode_share.sum(axis=1), axis=0)

mode_share_percent.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')
plt.title('Commute Mode Share by Trip Category')
plt.ylabel('Percentage of Trips')
plt.xlabel('Trip Category')
plt.legend(title='Mode', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

top_sov_tracts = tract_summary.sort_values(by='sov_share', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='sov_share', y=top_sov_tracts.index, data=top_sov_tracts, palette='Reds_r')
plt.title('Top 10 Origin Tracts by SOV Share (Intra-County Commutes)')
plt.xlabel('SOV Share')
plt.ylabel('Origin Tract FIPS')
plt.tight_layout()
plt.show()
```

```python
# Define alternative modes explicitly
alt_modes = ['walking', 'public_transit', 'biking']

# Group by origin tract and calculate totals
tract_summary = df_intra.groupby('origin_trct_fips_2020').agg(
    total_trips=('activity_id', 'count'),
    sov_trips=('primary_mode', lambda x: (x == 'private_auto').sum()),
    alt_trips=('primary_mode', lambda x: x.isin(alt_modes).sum())
)

# Calculate shares
tract_summary['sov_share'] = tract_summary['sov_trips'] / tract_summary['total_trips']
tract_summary['alt_mode_share'] = tract_summary['alt_trips'] / tract_summary['total_trips']


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(tract_summary['sov_share'], tract_summary['alt_mode_share'], alpha=0.7)
plt.axvline(0.75, color='red', linestyle='--', label='High SOV Threshold')
plt.axhline(0.10, color='green', linestyle='--', label='Low Alt Mode Threshold')
plt.xlabel('SOV Share')
plt.ylabel('Alt Mode Share')
plt.title('Tract-Level Commuter Patterns (Intra-County)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
opportunity_tracts = tract_summary[
    (tract_summary['sov_share'] > 0.75) & (tract_summary['alt_mode_share'] < 0.10)
].sort_values(by='sov_share', ascending=False).head(10)

opportunity_tracts[['sov_share', 'alt_mode_share']].plot(
    kind='barh', figsize=(10, 6), color=['#d62728', '#2ca02c']
)
plt.title('Top Opportunity Tracts: High SOV, Low Alt Mode')
plt.xlabel('Share of Trips')
plt.legend(['SOV Share', 'Alt Mode Share'])
plt.tight_layout()
plt.show()
```
